```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('HR_comma_sep.csv')
df.head()
df.dropna()
```

```python
y  = df.groupby('left')['left'].agg('count')
x = np.array([0,1])
```

```python
dummy1 = pd.get_dummies(df['salary'])
df= pd.concat([df,dummy1],axis='columns')
```

```python

df = df.drop(['last_evaluation','Department', 'promotion_last_5years','salary','average_montly_hours','Work_accident'], axis='columns')
```

```python
df
```

```python

condition1  = df['satisfaction_level'] > 0.7 
condition2 = df['left'] == 1
```

```python
df = df[~(condition1 & condition2)]
df
```

```python

condition3  = df['satisfaction_level'] < 0.3 
condition4 = df['left'] == 0
df = df[~(condition3 & condition4)]
df[df['left']==0]
```

```python
X = df.drop(['left'],axis ='columns')
y= df['left']
    
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, )

from sklearn.linear_model import LogisticRegression
reg_model = LogisticRegression()

reg_model.fit(X_train ,y_train)
reg_model.score(X_train ,y_train)
```
